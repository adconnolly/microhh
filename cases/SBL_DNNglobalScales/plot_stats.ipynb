```python
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt
```

```python
vars_profile = ['u','v','w','b']
zhmask = [0,0,1,0]
```

```python
stats = nc.Dataset('SBL1800.default.0000000.nc', 'r')
# stats = nc.Dataset('SBL1800.default.0007000.nc', 'r')
# stats = nc.Dataset('SBL1800.default.0000100.nc', 'r')
default = stats.groups['default']
thermo = stats.groups['thermo']

all_time = stats.variables['time']
start = 50
end = all_time.shape[0]
step = 10
t = stats.variables['time'][start:end]
z = stats.variables['z'][:]
zh = stats.variables['zh'][:]
zstart=0
zend=int(0.75*z.shape[0])
zend=z.shape[0]
zhend=zend
zhstart=zstart

for iv,varStr in enumerate(vars_profile):
    plt.figure(iv)
    try: 
        var=thermo.variables[varStr][start:end, :]
    except:
        var=default.variables[varStr][start:end, :]
    for n in range(0, end-start, step):
        if zhmask[iv]:
            plt.plot(var[n, zhstart:zhend], zh[zhstart:zhend],'.-',label=str(round(t[n])))
        else:
            plt.plot(var[n, zstart:zend], z[zstart:zend],'.-',label=str(round(t[n])))
        plt.xlabel(varStr)#r'$b [m^2 s^{-2}]$')
        plt.ylabel(r'$z [m]$')
        plt.legend()
```

```python
zstart=0
zend=5
zhend=zend
zhstart=zstart

for iv,varStr in enumerate(vars_profile[-1]):
    plt.figure(iv)
    try: 
        var=thermo.variables[varStr][start:end, :]
    except:
        var=default.variables[varStr][start:end, :]
    for n in range(0, end-start, step):
        if zhmask[iv]:
            plt.plot(var[n, zhstart:zhend], zh[zhstart:zhend],'.-',label=str(round(t[n])))
        else:
            plt.plot(var[n, zstart:zend], z[zstart:zend],'.-',label=str(round(t[n])))
        plt.xlabel(varStr)#r'$b [m^2 s^{-2}]$')
        plt.ylabel(r'$z [m]$')
        plt.legend()
```
