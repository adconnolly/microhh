```python
stats = nc.Dataset('SBL1800.default.0000000.nc', 'r')
default = stats.groups['default']
thermo = stats.groups['thermo']

all_time = stats.variables['time']
start = 0
end = all_time.shape[0]
step = 1
t = stats.variables['time'][start:end]
z = stats.variables['z'][:]
zh = stats.variables['zh'][:]
```

```python
vars_profile = ['u','v','w','b']
```

```python
for varStr in vars_profile:
    try: 
        var=thermo.variables[varStr][start:end, :]
    except:
        var=default.variables[varStr][start:end, :]
    for n in range(start, end, step):
        try:
            plt.plot(var[n, :], z)
        except:
            plt.plot(var[n, :], zh)
        plt.xlabel(varStr)#r'$b [m^2 s^{-2}]$')
        plt.ylabel(r'$z [m]$')
    plt.show()
```
